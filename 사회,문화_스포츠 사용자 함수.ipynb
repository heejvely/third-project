```python
def make_df(topic):
    
    import requests
    from bs4 import BeautifulSoup
    import pandas as pd

    time = []
    media = []
    title = []
    document = []

    headers = {"User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/90.0.4430.93 Safari/537.36"}

    if topic == '사회/문화':
        url = 'https://news.naver.com/main/main.naver?mode=LSD&mid=shm&sid1=103'
        req = requests.get(url, headers=headers)
        html = req.text
        soup = BeautifulSoup(html, 'html.parser')
        contents = soup.find_all(class_='cluster_group _cluster_content')
        for content in contents[:4]:
            link = content.find(class_='cluster_text_headline nclicks(cls_lif.clsart)')['href']
            req = requests.get(link, headers=headers)
            html = req.text
            soup = BeautifulSoup(html, 'html.parser')

            time.append(soup.find('span',class_='t11').text) # time
            media.append(content.find(class_='cluster_text_press').text) # media
            title.append(soup.title.text.split(':')[0]) # title
            document.append(soup.find(id='articleBodyContents').text.strip('\n동영상 뉴스\t')) # document

        df = pd.DataFrame({'time':time,'media':media,'title':title,'document':document})
               
    elif topic == '스포츠':
        url = 'https://sports.news.naver.com'
        req = requests.get(url, headers=headers)
        html = req.text
        soup = BeautifulSoup(html, 'html.parser')
        contents = soup.find_all(class_='today_item')

        for content in contents[:4]:
            link = url+content.find(class_='link_today')['href']  
            req = requests.get(link, headers=headers)
            html = req.text
            soup = BeautifulSoup(html, 'html.parser')

            time.append(soup.find(class_='info').text.split('최종수정')[0].split('기사입력')[1].strip()) # time
            media.append(content.find(class_='information').text.split('\n')[1]) # media
            title.append(content.find(class_='title').text) # title
            document.append(soup.find(id='newsEndContents').text.split('기사제공')[0].strip('\n')) # document

        df = pd.DataFrame({'time':time,'media':media,'title':title,'document':document})
        
    return df


make_df('사회/문화')
```
